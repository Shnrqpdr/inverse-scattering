```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_train = pd.read_csv('./dados_train_loss_runs_3.csv')
```

```python
df_train.head(5)
```

```python
df_cross_val = pd.read_csv('./dados_val_loss_runs_3.csv')
```

```python
for run in range(1,7):
    print(run)
```

```python
fig = plt.figure(figsize=(8,6))
axes = fig.add_axes([0.1,0.1,0.8,0.8])

for run in range(1, 7):
    axes.plot(df_train[f'run_{run}'], ls='-', lw=1.5, label=f'{run} hidden layer')

axes.set_title(f'Loss values of the train data for each hidden layer added')
axes.set_xlabel(f'epochs')
axes.legend(loc='upper right')
plt.grid(linestyle='-', linewidth=0.5)
plt.savefig('./backups/evaluateModel/01_08/loss_each_hidden_layer_train.png')
```

```python
fig = plt.figure(figsize=(8,6))
axes = fig.add_axes([0.1,0.1,0.8,0.8])

for run in range(1, 7):
    axes.plot(df_cross_val[f'run_{run}'], ls='-', lw=1.5, label=f'{run} hidden layer')

axes.set_title(f'Loss values of the cross-validation data for each hidden layer added')
axes.set_xlabel(f'epochs')
axes.legend(loc='upper right')
plt.grid(linestyle='-', linewidth=0.5)
plt.savefig('./backups/evaluateModel/01_08/loss_each_hidden_layer_val.png')
```
